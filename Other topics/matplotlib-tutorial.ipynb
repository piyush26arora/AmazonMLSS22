## MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

## Intro to pyplot
**matplotlib.pyplot** is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

[Matplotlib Pyplot Documentation Link](https://matplotlib.org/stable/tutorials/introductory/pyplot.html)

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)

plt.title('Sine-wave form ')
plt.plot(x,y)
```

```python
#Subplot
#Compute x and y coordinates for points on sine and cosine waves
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#Set up a subplot grid that has height 2 and width 1 ,
#and set the first such subplot as active.
plt.subplot(2,1,1)

#Make the first subplot
plt.plot(x,y_sin)
plt.title('Sine')

#Set the second subplot as active.
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r')
plt.title('Cosine')

#Show the plot (optional here)
plt.show()
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

# **Box Plot using Matplotlib**

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  
```

# **Pie Chart**

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.5, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
